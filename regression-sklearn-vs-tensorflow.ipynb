```python
#Importing the basic librarires for building model

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score,r2_score, mean_absolute_error, mean_squared_error, median_absolute_error



from sklearn.linear_model import LinearRegression 
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor

from sklearn.preprocessing import LabelEncoder # for make encoding / change from the category to numerical columns
```

```python
#show the data again
df.head()
```

```python
# separation the data type columns [ object and numeric ]

cat = []
num = []
for i in df.columns:
    if df[i].dtypes == 'O':
        cat.append(i)
    else:
        num.append(i)

data_n=df[num]   # new dataframe just type numeric
data_c=df[cat]  # new dataframe just type object
```

```python
# let start make the encoding 

label_encoders = {}
categorical_columns = data_c.columns  # I would recommend using columns names here if you're using pandas. If you're using numpy then stick with range(n) instead

for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    data_c[column] = label_encoders[column].fit_transform(data_c[column])
```

```python
# let show the new columns after encoding
data_c
```

```python
# merge 2 data set 
frames = [data_c, data_n]
  
data = pd.concat(frames,axis=1)
```

```python
#show the new dataframe after change the category columns
data.head()
```

```python
#show the data type - new dataframe
data.info()
```

- üßß All the columns / Data type numerical

### üëì  Now split the Data üé®

```python
#Defined X value and y value , and split the data train
X = data.drop(columns="Price")           
y = data["Price"]    # y = price

# split the data train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42) # split 25% test 75 % train

print("X Train : ", X_train.shape)
print("X Test  : ", X_test.shape)
print("Y Train : ", y_train.shape)
print("Y Test  : ", y_test.shape)
```

```python
#Defined object from library Regression  / from the libraries import before / 6 model

LR = LinearRegression()
DTR = DecisionTreeRegressor()
RFR = RandomForestRegressor()
KNR = KNeighborsRegressor()
MLP = MLPRegressor()
SVR = SVR()
```

```python
# make for loop for Regression 

li = [LR ,DTR ,RFR ,KNR ,MLP  ,SVR ]
score = []
for i in li:
    i.fit(X_train,y_train)   #traing the model
    ypred = i.predict(X_test) # start to predct
    score.append([i,r2_score(y_test,ypred)*100])
    
```

```python
Score_df = pd.DataFrame(score)
Score_df.columns =['Model', 'Score']
Score_df
```

# <p style="padding:10px;background-color:#d896ff;margin:0;color:#660066;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">4Ô∏è‚É£ Building Model using TF  - TensorFlow </p>

```python
#Importing the basic librarires for building model

import tensorflow as tf
import keras
```

```python
# model definition TF

Reg_M= keras.models.Sequential([
    
    keras.layers.Dense(8,activation='tanh'),
    keras.layers.Dense(128,activation="tanh"),
    keras.layers.Dense(64,activation="tanh"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1,activation='linear'), # we need linear model
    
])
```

```python
# make Compile 

Reg_M.compile(optimizer='adam',loss='mse',metrics=["mse"]) # mse = mean square error
```

```python
# fit the model

Fit_M = Reg_M.fit(X_train,y_train,
              validation_data=(X_test,y_test),
               epochs=1000,  # you can change the epochs and batch size based on the results
              batch_size=500, 
               verbose=1,
              callbacks=[ tf.keras.callbacks.EarlyStopping(
              patience=10,monitor='val_accuracy',restore_best_weights=True)])
```

### Attention ‚õî: 
- No accurese here,  but look the loss decreasing

```python
# The Summary 

Reg_M.summary()
```

```python
# Model Evaluate 

Model_loss, Model_Accuracy = Reg_M.evaluate(X_test,y_test)
print("Model Loss : " ,Model_loss )
print("Model Accuracy : " ,Model_Accuracy )
```

```python
# make the y predict 
y_pred = Reg_M.predict(X_test)
```

```python
# drow the loss 

plt.plot(Fit_M.history["loss"])
plt.plot(Fit_M.history["val_loss"])
plt.title("Model Loss")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.legend(["Train","Validation"])
plt.show()
```

```python
# print some results

print ("Mean Absolute Error : ", mean_absolute_error(y_test, y_pred))
print ("Mean Squared Error : ",mean_squared_error(y_test, y_pred))
print ("Median Absolute error : ",median_absolute_error(y_test, y_pred))
print ("r2 score : ",r2_score(y_test, y_pred))
```
